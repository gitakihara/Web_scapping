## WEB SCRAPPING

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {
    "User-Agent": "Mozilla/5.0"
}

records = []

# Adjust the number of pages as needed
for page in range(1, 21):

    if page == 1:
        url = "https://realtors.co.ke/property-type/apartments-for-sale/"
    else:
        url = f"https://realtors.co.ke/property-type/apartments-for-sale/page/{page}/"

    print(f"Scraping page {page}...")

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Failed on page {page}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")

    cards = soup.find_all(
        "div",
        class_=lambda x: x and "item-listing-wrap" in x
    )

    print(f"Found {len(cards)} properties")

    for card in cards:

        # TITLE + URL
        title = ""
        property_url = ""

        title_tag = card.find("h2", class_="item-title")

        if title_tag:
            link = title_tag.find("a")

            if link:
                title = link.get_text(strip=True)
                property_url = link.get("href")

        # LOCATION
        location = ""

        location_tag = card.find("address")

        if location_tag:
            location = location_tag.get_text(strip=True)

        # PRICE
        price = ""

        price_tag = card.find("span", class_="price")

        if price_tag:
            price = price_tag.get_text(strip=True)

        # Convert price to number
        numeric_price = None

        try:
            numeric_price = int(
                price.replace("Ksh", "")
                     .replace(",", "")
                     .strip()
            )
        except:
            pass

        records.append({
            "title": title,
            "price": price,
            "price_numeric": numeric_price,
            "location": location,
            "property_url": property_url
        })

    time.sleep(1)

# Create DataFrame
df = pd.DataFrame(records)

print("\nTotal Properties Scraped:", len(df))
print(df.head())

# Save CSV
df.to_csv("realtors_apartments.csv", index=False)

print("CSV saved successfully!")
```
